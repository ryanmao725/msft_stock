```python
# Time to define our gym environment where the baselines will run
import gym
from gym import spaces
```

```python
class StockEnv(gym.Env):
    def __init__(self, df: pd.DataFrame):
        super(StockEnv, self).__init__()
        # Set our df
        self.df = df

        # Set our action space
        self.action_space = spaces.Box(low=-1, high=1, shape=(1,))
        # Set our observation space
        # This will consist of:
        # the balance
        # the open of the current date
        # the amount of shares owned
        self.observation_space = spaces.Box(low=0, high=np.inf, shape=(3,))
        self.balance = 100000
        self.quantity = 0
        self.open = self.df.iloc(0)['Open']

        
```
