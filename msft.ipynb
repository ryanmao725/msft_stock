```python
# Conver the date column to datetime
msft_df['Date'] = pd.to_datetime(msft_df['Date'],format = '%m/%d/%Y %H:%M:%S')
display(msft_df.head())
display(msft_df.info())
display(msft_df.describe())
```

```python
# We don't care about the H:M:S part of the time, and we probably want to split the time up into days/years

msft_fe_df = msft_df.drop(labels = 'Date', axis = 1)
# msft_fe_df['Date'] = msft_fe_df['Date'].dt.normalize()
# msft_fe_df.insert(0, column = 'Day', value = msft_fe_df['Date'].dt.day)
# msft_fe_df.insert(0, column = 'Month', value = msft_fe_df['Date'].dt.month)
# msft_fe_df.insert(0, column = 'Year', value = msft_fe_df['Date'].dt.year)
# msft_fe_df.drop(labels = 'Date', axis = 1, inplace = True)

display(msft_fe_df.head())
display(msft_fe_df.info())
```

```python
# Split data into train, val and test groups
train, test = skms.train_test_split(msft_fe_df, train_size = 0.7, shuffle = False)
val, test = skms.train_test_split(test, train_size = 2/3, shuffle = False)

# Normalize
train_mean = train.mean()
train_std = train.std()

train = (train - train_mean) / train_std
val = (val - train_mean) / train_std
test = (test - train_mean) / train_std
```

```python
input_size = 8
shift_size = 1
predict_size = 1
window_size = input_size + shift_size

predict_label = 'High'

def _split(stack):
    inputs = stack[:, slice(0, input_size), :]
    labels = stack[:, slice(window_size - predict_size, None), :]
    labels = tf.stack(
        [labels[:, :, 4]],
        axis = -1
    )

    inputs.set_shape([None, input_size, None])
    labels.set_shape([None, predict_size, None])

    return inputs, labels

def make_dataset(data):
    data = np.array(data, dtype=np.float32)

    ds = tf.keras.preprocessing.timeseries_dataset_from_array(
        data=data,
        targets=None,
        sequence_length=window_size,
        sequence_stride=1,
        seed = 25,
        shuffle=True,
        batch_size=32,)

    # ds = ds.window(input_size, shift_size, True) 
    ds = ds.map(_split)

    return ds
```

```python
train_ds = make_dataset(train)
val_ds = make_dataset(val)
test_ds = make_dataset(test)

display(train_ds.element_spec)
# for inputs, outputs in train_ds.as_numpy_iterator():
#     print(f'{inputs} and {outputs}')
```

```python
def get_compiled_model(model):
    model.compile(
        loss = tf.losses.MeanSquaredError(),
        optimizer = tf.optimizers.Adamax(),
        metrics = [tf.metrics.MeanAbsoluteError()]
    )

    return model

early_stopping = tf.keras.callbacks.EarlyStopping(
    monitor = 'val_loss',
    mode = 'min',
    patience = 3,
)
```

## Multi-Layer Dense

```python
dense_model = tf.keras.Sequential([
    tf.keras.layers.Flatten(),
    tf.keras.layers.Dense(32, activation = 'relu'),
    tf.keras.layers.Dense(32, activation = 'relu'),
    tf.keras.layers.Dense(1),
    tf.keras.layers.Reshape([1, -1])
])

dense_model = get_compiled_model(dense_model)

dense_history = dense_model.fit(train_ds, epochs = 50, validation_data = val_ds, callbacks = [early_stopping])
```

```python
dense_predict = dense_model.predict(test_ds)
```

## CNN

```python
conv_model = tf.keras.Sequential([
    tf.keras.layers.Conv1D(filters=32,
                           kernel_size=(8,),
                           input_shape = (8, 5),
                           activation='relu'),
    tf.keras.layers.Dense(units=32, activation='relu'),
    tf.keras.layers.Dense(units=32, activation='relu'),
    tf.keras.layers.Dense(units=1),
])

conv_model = get_compiled_model(conv_model)

conv_history = conv_model.fit(train_ds, epochs = 50, validation_data = val_ds, callbacks = [early_stopping])
```

```python
conv_predict = conv_model.predict(test_ds)
```

## LSTM

```python
lstm_model = tf.keras.models.Sequential([
    tf.keras.layers.LSTM(32),
    # tf.keras.layers.LSTM(32),
    tf.keras.layers.Dense(units = 1)
])

lstm_model = get_compiled_model(lstm_model)

lstm_history = lstm_model.fit(train_ds, epochs = 50, validation_data = val_ds, callbacks = [early_stopping])
```

```python
lstm_predict = lstm_model.predict(test_ds)
```
